```python
# test merge lots of parquet files 
import pyarrow.parquet as pq
import glob
import time
# Record the start time
start_time = time.time()
files = glob.glob('/gpfs/data1/vclgp/xiongl/IS2global/result/X10_Y0/*.parquet')
print('Number of files:',len(files))
schema = pq.ParquetFile(files[0]).schema_arrow
with pq.ParquetWriter("/gpfs/data1/vclgp/xiongl/IS2global/result/X10_Y0.parquet", schema=schema) as writer:
    for file in files:
        writer.write_table(pq.read_table(file, schema=schema))
# Record the end time
end_time = time.time()
# Calculate the running time
running_time = end_time - start_time
print(f"Running time: {running_time} seconds")
```
